```python
import os
import shutil
from random import shuffle

def ensure_dir(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)


name_list = [f for f in os.listdir("/content/drive/My Drive/Data/train/images/")]
shuffle(name_list)

print(len(name_list))

dataset_len = len(name_list)
train_num = int(dataset_len * 0.8)
val_num = int(dataset_len * 0.1)
test_num = int(dataset_len * 0.1)

train_list = name_list[:train_num]
val_list = name_list[train_num:train_num + val_num]
test_list = name_list[train_num + val_num:]

for the_list in [train_list, val_list, test_list]:
  for fpath in the_list:
    shutil.copy("/content/drive/My Drive/Data/train/images/" + fpath, "/content/drive/My Drive/Data/split/images/" + fpath)


name_list = [f for f in os.listdir("/content/drive/My Drive/Data/train/labels/")]
shuffle(name_list)

print(len(name_list))

dataset_len = len(name_list)
train_num = int(dataset_len * 0.8)
val_num = int(dataset_len * 0.1)
test_num = int(dataset_len * 0.1)

train_list = name_list[:train_num]
val_list = name_list[train_num:train_num + val_num]
test_list = name_list[train_num + val_num:]

for the_list in [train_list, val_list, test_list]:
  for fpath in the_list:
    img_name_indl = fpath.rfind("/") + 1
    img_name_indr = fpath.rfind(".")
    fname = fpath[img_name_indl:img_name_indr]
    shutil.copy("/content/drive/My Drive/Data/train/labels/" + fpath, "/content/drive/My Drive/Data/split/labels/" + fname + ".txt")
```
